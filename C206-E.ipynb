# Redes Neuronales con Tensorflow y Keras

Tensorflow es un proyecto de código libre de Google para hacer inteligencia artificial con redes neuronales profundas en las siguientes variantes:

- Redes Neuronales Artificiales (AiNet / ANN) - Tienen próposito general y son usuadas para aprender `y = X` con una función de activación: `sigmoid`, `relu`, `softmax`, `tanh`, etc.
- Redes Neuronales Convolutivas (ConvNet / CNN) - Tienen próposito específico a entranmiento matricial como imágenes y video.
- Redes Neuronales Recurrentes (RNet / RNN) - Tienen un própisito específico a entrenamiento en vectores recurrentes como texto y audio.

Dentro de Tensorflow podemos usar optimizadores y tensores (también vectores y matrices), los cuales son matrices multidimensionales o hípervectores.

Tensorflow cuenta con un módulo llamado Keras que permite generar redes neuronales Densas, Convolutivas y Recurrentes.
