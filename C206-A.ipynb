```python
import numpy as np
import tensorflow as tf
from tensorflow import keras

x_train = [1, 2, 3]
y_train = [2, 4, 6]

print("x=", x_train)
print("y=", y_train)
```

```python
w = tf.Variable(tf.random.normal([1]), trainable = True, name='weight')
b = tf.Variable(tf.random.normal([1]), trainable = True, name='bias')

print("w=", w.numpy())
print("b=", b.numpy())
```

```python
@tf.function
def cost(w, b):
    y_out = w * x_train + b
    error = tf.reduce_mean(tf.square(y_train - y_out))
    return error
```

```python
cost(w, b)
```

```python
optimizer = tf.optimizers.SGD(learning_rate=0.01)

optimizer
```

```python
trainable_vars = [w, b]

epochs = 5 # Cuántas actualizaciones de `w` y `b`

for i in range(epochs):
    with tf.GradientTape() as tp:
        #your loss/cost function must always be contained within the gradient tape instantiation
        loss = cost(w, b)
    gradients = tp.gradient(loss, trainable_vars)
    print(f"i={i} w={w.numpy()} b={b.numpy()}")
    print(f"loss={loss}")
    print(f"    gw={gradients[0].numpy()} gb={gradients[1].numpy()}")

    optimizer.apply_gradients(zip(gradients, trainable_vars))

    tf.print(w)
    tf.print(b)

    print("-" * 70)

tf.print(w)
tf.print(b)
```

```python
x = [ [0, 0], [0, 1], [1, 0], [1, 1]]
y = [ [0], [1], [1], [0] ]

w = tf.Variable(tf.random.uniform((2, 1), minval=0.1 , maxval=0.9 , dtype=tf.float32), name='w')
b = tf.Variable(tf.zeros(1, dtype=tf.float32), name='b')

learning_rate = 0.1
epoch = 100

with tf.GradientTape(persistent=True) as tape:

  for i in range(epoch):
    debug = i % (epoch / 20) == 0

    if debug:
      print("-" * 40)
      print(f"EPOCH: {i}")
      print("-" * 40)

    y_out = tf.sigmoid(x @ w + b)
    loss = tf.reduce_mean(tf.square(y - y_out))

    [dl_dw, dl_db] = tape.gradient(loss, [w, b])

    if debug:
      print("x=", x)
      print("y=", y)

      print("-" * 40)

      print("w=", w.numpy())
      print("y=", b.numpy())

      print("-" * 40)

      print("y*=", y_out.numpy())

      print("loss=", loss.numpy())

      print("-" * 40)

      print("dl_dw=", dl_dw.numpy())
      print("dl_db=", dl_db.numpy())

      print("-" * 40)

    w = w - learning_rate * dl_dw
    b = b - learning_rate * dl_db

    if debug:
      print("w*=", w.numpy())
      print("b*=", b.numpy())

print("=" * 40)

print("w*=", w.numpy())
print("b*=", b.numpy())
```
