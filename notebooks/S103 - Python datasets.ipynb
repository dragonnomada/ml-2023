# S103 - Manipulación de datos con Python

Para manipular conjuntos de datos vamos a construir `datasets` generalmente usamos la librería de `pandas` que consiste en construir conjuntos de datos organizados en tablas o `Dataframes`.

Los datos generalmente contienen muestras (`samples`) que se componen de características (`features`) y se establecen principalmente por valores en columnas, campos o ejes de análisis.

Por ejemplo, podemos tener muestras de experimentos o mediciones obtenidas de datos reales.

| Fecha            | Temperatura | PH  | Masa |
|------------------|-------------|-----|------|
| 2023-01-01 15:45 | 34          | 6.7 | 3.4  |
| 2023-01-01 15:50 | 33.2        | 6.8 | 3.5  |
| 2023-01-01 16:01 | 28.4        | 6.9 | 3.8  |
| 2023-01-01 16:32 | 27.5        | 7.8 | 4.1  |
| 2023-01-01 17:15 | 27.8        | 7.6 | 4.5  |

La librería de Pandas nos permitirá manipular toda la información usando de fondo los arreglos n-dimensionales de Numpy y permitiéndonos hacer principalmente selección, filtrado, mapeo y reorganización de los datos.

```python
import pandas as pd
import seaborn as sns

df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv", sep=";")

df.columns
```

```python
sns.pairplot(df[["fixed acidity", "alcohol", "quality", "total sulfur dioxide"]])
```

```python
df["fixed acidity"]
```

```python
# Segmentación de datos
df[ df["fixed acidity"] < 4.5 ]
```

```python
# Segmentación doble
df[ df["fixed acidity"] < 4.5 ][ df["sulphates"] < 0.4 ]
```

```python
# Consultas sobre el Dataframe
df.query(" `fixed acidity` < 4.5 and `sulphates` < 0.4 ")
```

```python
100 * df.query(" `fixed acidity` < 6 ").count()[0] / df.count()[0]
```
