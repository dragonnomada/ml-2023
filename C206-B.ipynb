```python
from tensorflow import keras

# x = [ (x1, x2) ] -> 2 inputs
x = [ [0, 1], [1, 0], [0, 0], [1, 1], [0, 1] ]
# y = [ y1 ] -> 1 output
y = [ 1, 1, 0, 0, 1]

model = keras.Sequential(
    [
        # Input
        keras.Input(shape=(2,)),
        # H1
        keras.layers.Dense(3, activation="sigmoid", name="Dh1"),
        # H2
        keras.layers.Dense(2, activation="sigmoid", name="Dh2"),
        # Output
        keras.layers.Dense(1, name="Dout"),
    ]
)

model.build()
```

```python
model.summary()
```

```python
import tensorflow as tf

model.compile('sgd', tf.keras.losses.MeanSquaredError(reduction="auto", name="mean_squared_error"))

model.fit(x, y)
```

```python
model.predict([[0, 0], [0, 1], [1, 0], [1, 1], [0.3, 0.7]])
```
